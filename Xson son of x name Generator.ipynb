```python
import opendatasets as od
import pandas as pd
```

```python
names = pd.read_csv('brazilian-names.csv')
```

```python
names = names.drop(columns=['Masc'])
names = names.rename(columns = {'Abel': 'Name:'}, inplace = False)
```

```python
names = names[names['Name:'].str.contains("son")]
```

```python
names['Surname:'] = names['Name:'].map(lambda x: 'Son of ' + x[:-3].strip().replace(" ",""))
```
